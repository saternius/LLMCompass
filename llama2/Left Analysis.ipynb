```python
import simplejson as json
```

```python
results = []
with open("leftorright.jsonl") as f:
    for line in f:
        obj = json.loads(line)
        positions = {}
        for x in obj['results']:
            positions[x[0]] = x[1]

        results.append({
            'idx': obj['idx'],
            'question': obj['question'],
            'positions': positions
        })
```

```python
questions = json.load(open("10group.json"))['questions']
quest_map = {q['question']:q for q in questions}
```

```python
compiled = []
for r in results:
    question = r['question']
    position = r['positions']
    print(question)
    print(position)
    
    obj = r.copy()
    obj['effect'] = {}
    q = quest_map[question]
    
    for e in q['effects']:
        obj['effect'][notation[e][1]] = q['effects'][e]
        
    compiled.append(obj)
    print(q)
    print("\n\n")
```

```python
compiled.sort(key=lambda x:x['idx'])
```

```python
outfile = open("left_right_10group.json", "w")
outfile.write(json.dumps({'results':compiled}))
outfile.close()
```

```python
left_score = {}
left_weight = {}

right_score = {}
right_weight = {}

for c in compiled:
    rw = c['positions']['right wing']
    lw = c['positions']['left wing']
    for e in c['effect']:
        if e not in left_score:
            left_score[e] = 0
            left_weight[e] = 0
            right_score[e] = 0
            right_weight[e] = 0
        
        d = c['effect'][e]
        
        left_sign = 1 if lw > rw else -1
        right_sign = 1 if rw > lw else -1
        
        left_score[e] += left_sign*d*ae_bias[e]
        left_weight[e] += left_sign*abs(d)*ae_bias[e]
        right_score[e] += right_sign*d*ae_bias[e]
        right_weight[e] += right_sign*abs(d)*ae_bias[e]
```
