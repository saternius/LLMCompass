```python
import simplejson as json
from IPython.core.display import display, HTML
from selenium import webdriver
```

```python
results = []
dists = []
with open('results/leftorright-vividmind.jsonl') as f:
    for line in f:
        obj = json.loads(line)
        positions = {}
        for x in obj['results']:
            positions[x[0]] = x[1]

        results.append({
            'idx': obj['idx'],
            'question': obj['question'],
            'positions': positions
        })
        dist = abs(positions['left wing'] - positions['right wing'])
        dists.append([obj['question'], dist, positions['left wing'], positions['right wing']])
results
```

```python
outfile = open("results/leftright_vividmind.json", "w")
outfile.write(json.dumps({'results':results}))
outfile.close()
```

```python
left_rows = results.copy()
right_rows = results.copy()
left_rows.sort(key=lambda x:x['positions']["left wing"], reverse=True)
right_rows.sort(key=lambda x:x['positions']["right wing"], reverse=True)
```

```python

print("{0:73}{1:65}".format("\033[1mMost left wing statements:\033[0m","\033[1mMost right wing statements:\033[0m"))
for i in range(25):
    ltxt = left_rows[i]['question']
    rtxt = right_rows[i]['question']
    print("{0:65}{1:65}".format(ltxt,rtxt))
```
