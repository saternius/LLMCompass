```python
import simplejson as json
```

```python
results = []
with open('results/leftorright-elephantbig5.jsonl') as f:
    for line in f:
        obj = json.loads(line)
        positions = {}
        for x in obj['results']:
            positions[x[0]] = x[1]

        results.append({
            'idx': obj['idx'],
            'question': obj['question'],
            'positions': positions
        })
results
```

```python
big5 = json.load(open("data/elephantcastle-big5/testdata.json"))
```

```python
quest_map = {}
for b in big5:
    d = b['text']['en'].lower()
    if d[:2] == "i ":
        d = d[2:]
    txt = "I "+d+"."
    quest_map[txt] = {
        'facet': b['facet'],
        'R': b['R']
    }
```

```python
quest_map['I tend to vote for liberal political candidates.']
```

```python
quest_map['I tend to vote for conservative political candidates.']
```

```python
rw_score = {}
lw_score = {}
compiled = []
dists = []
for r in results:
    q = quest_map[r['question']]
    lw = r['positions']['left wing']
    rw = r['positions']['right wing']
    tag = q['facet']+"-"+q['R']
    if tag not in lw_score:
        lw_score[tag] = 0
        rw_score[tag] = 0
        
    lw_score[tag] += lw
    rw_score[tag] += rw
    obj = q.copy()
    q['question'] = r['question']
    obj['positions'] = {
        'left': lw,
        'right': rw
    }
    dist = abs(lw - rw)
    dists.append([dist, r['question'], lw, rw])
    compiled.append(obj)
```

```python
outfile = open("big5_elephant.json", "w")
outfile.write(json.dumps({'results':compiled}))
outfile.close()
```

```python
summed_lw = {}
summed_rw = {}
summed_dists = []

l_reduced = {'O':0,'C':0,'E':0,'A':0,'N':0}
r_reduced = {'O':0,'C':0,'E':0,'A':0,'N':0}
d_reduced = {'O':0,'C':0,'E':0,'A':0,'N':0}

for a in atts:
    lf = lw_score[a+"-F"] if a+"-F" in lw_score else 0
    rf = rw_score[a+"-F"] if a+"-F" in rw_score else 0
    lr = lw_score[a+"-R"] if a+"-R" in lw_score else 0
    rr = rw_score[a+"-R"] if a+"-R" in rw_score else 0
    
    ls = lf-lr
    rs = rf-rr
    
    dist = (ls - rs)
    
    summed_lw[a] = ([ls, notation[a], a])
    summed_rw[a] = ([rs, notation[a], a])
    summed_dists.append([dist, notation[a], a, ls, rs])
    
    prop = a[0]
    l_reduced[prop] += ls
    r_reduced[prop] += rs
    d_reduced[prop] += dist
    
# summed_lw.sort(key=lambda x:x[0], reverse=True)
# summed_rw.sort(key=lambda x:x[0], reverse=True)
# summed_dists.sort(key=lambda x:x[0], reverse=True)

# print("LEFT: ")
# for s in summed_lw:
#     print(s)
# print("\n")
# print("RIGHT: ")
# for s in summed_rw:
#     print(s)
    
# print("\n")
# print("BIGGEST DIST: ")
# for s in summed_dists:
#     print(s)
```
