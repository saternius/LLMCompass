```python
import simplejson as json
```

```python
results = []
with open('results/leftorright-elephantbig5.jsonl') as f:
    for line in f:
        obj = json.loads(line)
        positions = {}
        for x in obj['results']:
            positions[x[0]] = x[1]

        results.append({
            'idx': obj['idx'],
            'question': obj['question'],
            'positions': positions
        })
results
```

```python
position = "left wing"
results.sort(key=lambda x:x['positions'][position], reverse=True)
for r in results:
    print("{0}\n{1}\n\n".format(r['question'],r['positions'][position]))
```

```python
big5 = json.load(open("data/elephantcastle-big5/testdata.json"))
```

```python
quest_map = {}
for b in big5:
    d = b['text']['en'].lower()
    if d[:2] == "i ":
        d = d[2:]
    txt = "I "+d+"."
    quest_map[txt] = {
        'facet': b['facet'],
        'R': b['R']
    }
```

```python
quest_map['I tend to vote for liberal political candidates.']
```

```python
quest_map['I tend to vote for conservative political candidates.']
```

```python
rw_score = {}
lw_score = {}
compiled = []
for r in results:
    q = quest_map[r['question']]
    lw = r['positions']['left wing']
    rw = r['positions']['right wing']
    tag = q['facet']+"-"+q['R']
    if tag not in lw_score:
        lw_score[tag] = 0
        rw_score[tag] = 0
        
    lw_score[tag] += lw
    rw_score[tag] += rw
    obj = q.copy()
    q['question'] = r['question']
    obj['positions'] = {
        'left': lw,
        'right': rw
    }
    compiled.append(obj)
```

```python
outfile = open("big5_elephant.json", "w")
outfile.write(json.dumps({'results':compiled}))
outfile.close()
```
